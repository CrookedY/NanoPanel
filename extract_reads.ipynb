```python
os.chdir('/home/matthew/Documents/hackmed18/reads')
with open('../Bham_Run1_20171009_DirectRNA.files.txt') as f:
    for fast5 in f:
        fast5 = fast5.strip()
        !wget {fast5} 2> /dev/null
        fast5 = os.path.split(fast5)[1]
        f = h5py.File(fast5)
        read = list(f['Raw/Reads'].keys())[0]
        read_id = f['Raw/Reads/{}'.format(read)].attrs['read_id'].decode()
        f.close()
        !mv {fast5} {read_id}.fast5
```
